```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("spotify_tracks.csv")
core_features = ['valence', 'energy', 'danceability', 'acousticness', 'tempo']

# Extract columns
columns = [
    'track_name', 'artists', 'album_name', 'track_id', 
    'valence', 'energy', 'danceability', 'acousticness', 'tempo', 'track_genre'
]
# Clean data
df_clean = df[columns].copy().dropna()
df_clean = df_clean.drop_duplicates(subset=['track_name', 'artists'], keep='first')

# Scale audio features between 0.0 and 1.0
scaler = MinMaxScaler()
df_clean[core_features] = scaler.fit_transform(df_clean[core_features])

# Results
print("First 5 rows of scaled data:")
print(df_clean.head())

df_clean.to_csv('processed_spotify_data.csv', index=False)
print("\nTotal rows after cleaning:", len(df_clean))
```
